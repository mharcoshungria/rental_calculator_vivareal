```python
import pandas as pd
df = pd.read_csv('dados_apartamentos.csv')
```

```python
df.head(40)
```

```python
df = df.drop(['Unnamed: 0', 'data_scraping'],axis=1)
```

```python
df['vagas'] = df['vagas'].str.extract('(.*) Vaga')
df['banheiros'] = df['banheiros'].str.extract('(.*) Banheiro')
df['quartos'] = df['quartos'].str.extract('(.*) Quarto')
df.head()
```

```python
df['preco'] = df['preco'].str.extract('R\$ (.*) /mês') 
df['condominio'] = df['condominio'].str.extract('R\$ (.*)')
df.head()
```

```python
df['preco'] = df['preco'].replace('\.','',regex=True)
df['condominio'] = df['condominio'].replace('\.','',regex=True)
df.head()
```

```python
df_sp = df[df.cidade != 'Rio de Janeiro']
```

```python
# Se quiser excluir um dataframe, rode del <dataframe>
```

```python
df_sp.head()
```

```python
df_sp = df_sp.drop(['cidade'],axis=1)
df_sp.head()
```

```python
df_sp['bairro'] = df_sp['endereco'].str.split('\-').str[1]
df_sp['bairro'] = df_sp['bairro'].str.split('\,').str[0]
df_sp.head()
```

```python
df_sp['endereco'] = df_sp['endereco'].str.split('\-').str[0]
df_sp.head()
```

```python
df_sp['vagas'] = df_sp['vagas'].replace('--','0',regex=True)
df_sp.head()
```

```python
df_sp['vagas'].unique()
```

```python
df_sp['bairro'].unique()
```

```python
df_sp.dtypes
```

```python
cols = ['quartos','vagas','preco','condominio','banheiros']
df_sp[cols] = df_sp[cols].apply(pd.to_numeric, errors='coerce')
```

```python
df_sp.dtypes
```

```python
df_sp.describe()
```

```python
df_sp.isnull().sum()
```

```python
df_sp.info()
```

```python
df_sp = df_sp.dropna(subset=['preco'])
```

```python
df_sp = df_sp.fillna(0)
```
